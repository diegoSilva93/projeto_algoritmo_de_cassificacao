```python
#Download bibliotecas
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#carregamento do dataset

dataset = pd.read_csv("Churn_Modelling.csv.csv")
```

```python
#Fazendo anÃ¡lise exploratÃ³ria dos dados
dataset
```

```python
# Selecionando variÃ¡veis para X

X = dataset.iloc[:, 3:13].values
```

```python
# Selecionand variavel dependente Y
Y = dataset.iloc[:, 13].values
```

```python
#Transformando dados categÃ³ricos em numÃ©ricos
label_encoder_X1 = LabelEncoder()
X[:,1] = label_encoder_X1.fit_transform(X[:,1])

label_encoder_X2 = LabelEncoder()
X[:,2] = label_encoder_X1.fit_transform(X[:,2])
```

```python
one_hot_encoder = make_column_transformer((OneHotEncoder(categories = "auto", sparse = False), [1]), remainder = "passthrough")
```

```python
X = one_hot_encoder.fit_transform(X)
X = X[:,1:]
```

```python
# Treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = Sequential()
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, Y_train, batch_size = 10, epochs = 100)

Y_pred = classifier.predict(X_test)
Y_pred = (Y_pred > 0.5)

cm = confusion_matrix(Y_test, Y_pred)
```

```python
print(cm)
```

```python
acuracia = accuracy_score(Y_test,Y_pred)
```

```python
print(acuracia)
```
